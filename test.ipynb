### Prepare to test a baseline method on Salinas

```python
import numpy as np
import sklearn.svm
import seaborn as sns
```

```python
N_CLASSES = len(salinas_label_values)
N_BANDS = salinas_img.shape[-1]
```

Generate a color palette that can be used to visualize the datasets

```python
if salinas_palette is None:
    # Generate color palette
    salinas_palette = {0: (0, 0, 0)}
    for k, color in enumerate(sns.color_palette("hls", len(salinas_label_values) - 1)):
        salinas_palette[k + 1] = tuple(np.asarray(255 * np.array(color), dtype='uint8'))
invert_palette = {v: k for k, v in salinas_palette.items()}

def convert_to_color(x):
    return utils.convert_to_color_(x, palette=salinas_palette)
def convert_from_color(x):
    return utils.convert_from_color_(x, palette=invert_palette)
```

Sample the training and test datasets from the original dataset and display these

```python
SAMPLE_PERCENTAGE = 0.5
SAMPLING_MODE = 'random' #random, fixed, disjoint

train_gt, test_gt = utils.sample_gt(salinas_gt, SAMPLE_PERCENTAGE, mode=SAMPLING_MODE)
print("{} samples selected (over {})".format(np.count_nonzero(train_gt), np.count_nonzero(salinas_gt)))

utils.display_predictions(convert_to_color(train_gt), vis, caption="Train ground truth")
utils.display_predictions(convert_to_color(test_gt), vis, caption="Test ground truth")
```

#### Run the training of the SVM baseline method from sklearn on the generated training and test datasets

```python
CLASS_BALANCING = False

X_train, y_train = utils.build_dataset(salinas_img, train_gt, ignored_labels=salinas_ignored_labels)
class_weight = 'balanced' if CLASS_BALANCING else None
clf = sklearn.svm.SVC(class_weight=class_weight)
clf.fit(X_train, y_train)
#save_model(clf, 'SVM', 'Salinas')
prediction = clf.predict(salinas_img.reshape(-1, N_BANDS))
prediction = prediction.reshape(salinas_img.shape[:2])
```

Display the results

```python
run_results = utils.metrics(prediction, test_gt, ignored_labels=salinas_ignored_labels, n_classes=N_CLASSES)

mask = np.zeros(salinas_gt.shape, dtype='bool')
for l in salinas_ignored_labels:
    mask[salinas_gt == l] = True
prediction[mask] = 0

color_prediction = convert_to_color(prediction)
utils.display_predictions(color_prediction, vis, gt=convert_to_color(test_gt), caption="Prediction vs. test ground truth")

utils.show_results(run_results, vis, label_values=salinas_label_values)
```

#### Run another baseline algorithm to test performance, this time test Label Propagation from sklearn

```python
from sklearn.semi_supervised import LabelSpreading
```

Divide set into labeled and unlabeled dataset. Here we can use the test dataset as the unlabeled dataset, put these labels to -1. It should be possible to use these for training since we don't use the labels, while we also can use them for testing the performance later.

```python
X_labeled, y_labeled = utils.build_dataset(salinas_img, train_gt, ignored_labels=salinas_ignored_labels)
X_unlabeled, y_unlabeled = utils.build_dataset(salinas_img, test_gt, ignored_labels=salinas_ignored_labels)

y_labeled = np.int8(y_labeled.tolist())
y_unlabeled = np.int8(y_unlabeled.tolist())

y_unlabeled_saved = np.copy(y_unlabeled)
y_unlabeled[:] = -1


X_train = np.concatenate((X_labeled, X_unlabeled))
y_train = np.concatenate((y_labeled, y_unlabeled))
```

```python
y_unlabeled_saved
```

Run Label Spreading on the new dataset with k-nearest neighbors. Alpha in this case is the clamping factor, this decides how much a sample should rely on its initial label or information from its neighbors.

```python
k = 10
a = 0.5

label_spread_model = LabelSpreading(kernel='knn', n_neighbors=k, alpha = a, max_iter=100)
fit = label_spread_model.fit(X_train, y_train)
labeled_score = fit.score(X_labeled, y_labeled)
unlabaled_score = fit.score(X_unlabeled, y_unlabeled_saved)
```

Error with division by zero means that there are clusters that have no labeled sample connected to them. This is a known issue, increase the amount of neighbors for example.

```python
print('LabelSpread', 'knn', k, a, labeled_score, unlabaled_score)
```

```python
y_pred = label_spread_model.predict(salinas_img.reshape(-1, N_BANDS))
```

```python
y_pred = y_pred.reshape(salinas_img.shape[:2])
```

```python
run_results = utils.metrics(y_pred, test_gt, ignored_labels=salinas_ignored_labels, n_classes=N_CLASSES)

mask = np.zeros(salinas_gt.shape, dtype='bool')
for l in salinas_ignored_labels:
    mask[salinas_gt == l] = True
y_pred[mask] = 0

color_prediction = convert_to_color(y_pred)
utils.display_predictions(color_prediction, vis, gt=convert_to_color(test_gt), caption="Prediction vs. test ground truth")

utils.show_results(run_results, vis, label_values=salinas_label_values)
```
