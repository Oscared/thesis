```python
import numpy as np
import sklearn.model_selection
import utils
import visdom
from datasets import get_dataset, HyperX
import seaborn as sns
from sklearn.decomposition import PCA
import matplotlib.pyplot as plt
```

```python
vis = visdom.Visdom()
salinas_img, salinas_gt, salinas_label_values, salinas_ignored_labels, salinas_rgb_bands, salinas_palette = get_dataset("Salinas", target_folder='/Data/')
N_CLASSES = len(salinas_label_values)
N_BANDS = salinas_img.shape[-1]

if salinas_palette is None:
    # Generate color palette
    salinas_palette = {0: (0, 0, 0)}
    for k, color in enumerate(sns.color_palette("hls", len(salinas_label_values) - 1)):
        salinas_palette[k + 1] = tuple(np.asarray(255 * np.array(color), dtype='uint8'))
invert_palette = {v: k for k, v in salinas_palette.items()}

def convert_to_color(x):
    return utils.convert_to_color_(x, palette=salinas_palette)
def convert_from_color(x):
    return utils.convert_from_color_(x, palette=invert_palette)
```

```python
part_1, part_2 = utils.sample_gt(salinas_gt, 0.1, mode='disjoint')
utils.display_predictions(convert_to_color(part_1), vis, caption="Part 1 ground truth")
utils.display_predictions(convert_to_color(part_2), vis, caption="Part 2 ground truth")
```

```python
part_3, part_4 = utils.sample_gt(part_1, 0.1, mode='disjoint')
utils.display_predictions(convert_to_color(part_3), vis, caption="Part 3 ground truth")
utils.display_predictions(convert_to_color(part_4), vis, caption="Part 4 ground truth")
```

```python
#Do PCA data aug. Sklearn PCA then augmentation?
centered_img = salinas_img - np.mean(salinas_img)
data_train, target_train = utils.build_dataset(centered_img, part_3, ignored_labels = salinas_ignored_labels)

pca = PCA(n_components=11)
print(np.shape(data_train))
pca.fit(data_train)

print(pca.explained_variance_)
print(pca.n_features_)
print(pca.n_samples_)

print(np.shape(data_train[0,:].reshape(1,-1)))

data=salinas_img[0:5,0:5]

data_aug = np.zeros_like(data)

for idx, _ in np.ndenumerate(salinas_gt[0:5,0:5]):
    print('Loop')
    print(idx)
    print(np.shape(data))
    print(np.shape(data[idx, :]))
    data_pca = pca.transform(data[idx, :].reshape(1,-1))

    alpha = np.random.uniform(0.9,1.1)
    print(alpha)

    data_pca[:,0] = data_pca[:,0]*alpha
    
    print(np.shape(data_pca))

    data_aug[idx,:] = pca.inverse_transform(data_pca)
    print(np.shape(data_aug))
    print(np.shape(data_aug[idx,:]))
    
```

```python
mask = np.ones_like(salinas_gt)
mask[salinas_gt == 0] = 0
x_pos, y_pos = np.nonzero(mask)
indices = np.array([(x,y) for x,y in zip(x_pos, y_pos)])
labels = [salinas_gt[x,y] for x,y in indices]
np.random.shuffle(indices)

class_var = {}

for c in np.unique(salinas_gt):
    if c not in salinas_ignored_labels:
        l_indices = np.nonzero(labels==c)
        pos = indices[l_indices]
        data = salinas_img[pos[:,0], pos[:,1]]
        print(np.shape(data))
        var = np.var(data, axis=0)
        diag_var = np.diag(var)
        class_var[c] = diag_var

diag_var = class_var[1]

noise = np.random.multivariate_normal(np.zeros(np.shape(diag_var)[0]), diag_var)
```
